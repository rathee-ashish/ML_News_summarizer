```python
import io
import webbrowser
import requests
from tkinter import *
from urllib.request import urlopen
from PIL import ImageTk,Image
import xmltodict,json
from bs4 import BeautifulSoup

H = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
#function to get response of the links
def function (url):
    response=requests.get(url,headers=H)
    soup = BeautifulSoup(response.text, "html.parser")
    return soup

    
```

```python
#opening api into xml file
from urllib.request import urlopen
from xml.etree.ElementTree import parse

var_url = urlopen('https://news.google.com/news/rss')
xmldoc = parse(var_url)
```

```python
#extracting links from api
url=[]
text=[]
for item in xmldoc.iterfind('channel/item'):
    #title = item.findtext('title')
    #date = item.findtext('pubDate')
    link = item.findtext('link')
    #a=requests.head(link,allow_redirects=True).url
    #print(a)
    htmlpage=function(link)
    redirected_link = htmlpage.get_text()
    redirected_link=redirected_link.replace("Google NewsOpening ","")   
    
    htmlpage_r=function(redirected_link)
    
    text_elements = htmlpage_r.find_all("p")
    text.append(" ".join([element.text for element in text_elements]))
    
    url.append(redirected_link)
    
    
    
```
